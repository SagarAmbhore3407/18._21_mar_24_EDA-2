```python
#Q1. What is the average rating for each product category?


import pandas as pd

# Load the dataset
file_path = 'amazon.csv'
df = pd.read_csv(file_path)


df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# average rating for each product category
average_rating_per_category = df.groupby('category')['rating'].mean().reset_index()

# all rows of the calculated average ratings per category
print(average_rating_per_category)


```

```python

#Q2.  What are the top rating_count products by category?


# Convert 'rating_count' to numeric, handling commas
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(float, errors='ignore')

# Get the top rating_count products by category
top_rating_count_products = df.loc[df.groupby('category')['rating_count'].idxmax()].reset_index(drop=True)

# Display the top rating_count products by category
print(top_rating_count_products[['category', 'product_name', 'rating_count']])
```

```python
#Q3.  What is the distribution of discounted prices vs. actual prices?

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
# for actual prices
plt.subplot(1, 2, 1)
sns.histplot(df['actual_price'], bins=30, kde=True)
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')

# for discounted prices
plt.subplot(1, 2, 2)
sns.histplot(df['discounted_price'], bins=30, kde=True)
plt.title('Distribution of Discounted Prices')
plt.xlabel('Discounted Price')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python

#Q5. What are the most popular product names?

import pandas as pd


file_path = 'amazon.csv'  
amazon_data = pd.read_csv(file_path)



amazon_data['discount_percentage'] = amazon_data['discount_percentage'].str.replace('%', '').astype(float)

average_discount_by_category = amazon_data.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)

# Display the result
print(average_discount_by_category)

amazon_data['rating_count'] = amazon_data['rating_count'].fillna('0').str.replace(',', '').astype(int)

most_popular_products = amazon_data.groupby('product_name')['rating_count'].sum().sort_values(ascending=False)

# Display the top 10 most popular products
top_10_popular_products = most_popular_products.head(10)
print(top_10_popular_products)

```

```python

#Q6. What are the most popular product keywords?

from collections import Counter
import re


stop_words = set([
    'a', 'an', 'and', 'are', 'as', 'at', 'be', 'by', 'for', 'from', 'has', 'he',
    'in', 'is', 'it', 'its', 'of', 'on', 'that', 'the', 'to', 'was', 'were', 
    'will', 'with', 'this', 'these', 'those', 'which', 'i', 'you', 'we', 'they'
])


def clean_tokenize(text):
    # to lowercase
    text = text.lower()
    # Remove non-alphanumeric
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Split into words
    words = text.split()
    # Remove stop words
    filtered_words = [word for word in words if word not in stop_words]
    return filtered_words


combined_text = (amazon_data['product_name'].astype(str) + " " + amazon_data['about_product'].astype(str)).str.cat(sep=' ')

# Tokenize the combined text
tokens = clean_tokenize(combined_text)


keyword_counts = Counter(tokens)


top_10_keywords = keyword_counts.most_common(10)
print(top_10_keywords)
```

```python
#Q7. What are the most popular product reviews?


# Replace NaN values with 0 in 'rating_count' and convert to integer after removing commas
amazon_data['rating_count'] = amazon_data['rating_count'].fillna('0').str.replace(',', '').astype(int)

# Sort the data by 'rating_count' and 'rating' in descending order
sorted_reviews = amazon_data.sort_values(by=['rating_count', 'rating'], ascending=[False, False])

# Select the top 10 most popular reviews
top_10_popular_reviews = sorted_reviews[['product_name', 'review_title', 'review_content', 'rating', 'rating_count']].head(10)

# Display the top 10 most popular reviews
print(top_10_popular_reviews)
```

```python
#Q8. What is the correlation between discounted_price and rating?



df['discounted_price'] = df['discounted_price'].replace('[,]', '', regex=True).astype(float)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate the correlation between discounted_price and rating
correlation = df['discounted_price'].corr(df['rating'])
print("Correlation between discounted_price and rating:", correlation)
```

```python
#Q9. What are the Top 5 categories based on the highest ratings?



df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Group by 'category' and calculate the mean rating for each category
category_ratings = df.groupby('category')['rating'].mean()

# Sort the categories based on the mean rating in descending order and get the top 5
top_5_categories = category_ratings.sort_values(ascending=False).head(5)
print("Top 5 categories based on highest ratings:")
print(top_5_categories)
```

```python
#Q10. Identify any potential areas for improvement or optimization based on the data analysis.



''' 
1. Data Quality and Completeness:
Missing Values: Check for missing values in important columns such as rating, discounted_price, and rating_count. 
Fill or handle these missing values appropriately.
Data Consistency: Ensure that all price-related columns are consistently formatted and that all ratings are within the expected range (1-5).

2. Product Performance Analysis:
Low-Rated Products: Identify products with consistently low ratings. Investigate the reasons behind the low ratings by analyzing the reviews.
High-Rated Products: Identify products with high ratings and analyze what makes them successful. Use these insights to improve other products.

3.Category Performance:
Top Performing Categories: Focus on categories with the highest average ratings and explore opportunities to introduce new products or 
enhance existing ones in these categories.
Underperforming Categories: Identify categories with low average ratings and analyze the reasons. Consider 
improving product quality or adjusting marketing strategies.

4.Price Optimization:
Price vs. Rating: Analyze the relationship between price and rating to find the optimal price points that maximize both sales and customer satisfaction.
Discount Analysis: Evaluate the effectiveness of discounts by comparing the discounted price with ratings and sales volume. 
Optimize discount strategies to boost sales while maintaining profitability.

5. Review Analysis:
Sentiment Analysis: Perform sentiment analysis on review content to identify common positive and negative themes. 
Use this information to address customer concerns and improve product features.

6.Customer Engagement:
Active Reviewers: Identify and engage with active reviewers to gather more detailed 
feedback and build a loyal customer base.
Review Response: Consider responding to customer reviews, both positive and negative, 
to show that you value customer feedback and are committed to improving their experience.

'''
from textblob import TextBlob
from collections import Counter


missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values,"\n")

# Drop rows with missing values in 'rating', 'discounted_price', and 'rating_count'
df = df.dropna(subset=['rating', 'discounted_price', 'rating_count'])
# Convert 'rating' to numeric values
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Identifying low-rated productss.
low_rated_products = df[df['rating'] < 3]
print("Low-rated products:")
print(low_rated_products[['product_id', 'product_name', 'rating', 'review_content']],"\n")



# sentiment analysis on review content
df['sentiment'] = df['review_content'].apply(lambda x: TextBlob(x).sentiment.polarity)
print("Sentiment analysis results:")
print(df[['product_id', 'product_name', 'rating', 'sentiment']],"\n")



# Convert 'discounted_price' to numeric values
df['discounted_price'] = df['discounted_price'].replace('[,]', '', regex=True).astype(float)
# relationship between price and rating
price_rating_corr = df['discounted_price'].corr(df['rating'])
print("Correlation between discounted price and rating:", price_rating_corr,"\n")

# Group by 'category' and calculate the mean rating for each category
category_ratings = df.groupby('category')['rating'].mean()

# categories based on the mean rating in descending order and get the top 5
top_5_categories = category_ratings.sort_values(ascending=False).head(5)
print("Top 5 categories based on highest ratings:")
print(top_5_categories,"\n")

# Identify categories with low average ratings
bottom_5_categories = category_ratings.sort_values().head(5)
print("Bottom 5 categories based on lowest ratings:")
print(bottom_5_categories,"\n")



# Identify frequent issues mentioned in negative reviews
negative_reviews = df[df['rating'] < 3]['review_content']
negative_words = ' '.join(negative_reviews).split()
common_issues = Counter(negative_words).most_common(10)
print("Common issues in negative reviews:")
print(common_issues)
```
