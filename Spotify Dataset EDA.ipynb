```python
#Q1.Load the dataframe and ensure data quality by checking for missing values and duplicate rows.
# Handle missing values and remove duplicate rows if necessary.

import pandas as pd

# Load the dataset
file_path = 'spotify.csv'
spotify_data = pd.read_csv(file_path)

# Check for missing values
missing_values = spotify_data.isnull().sum()
print("Missing values per column:\n", missing_values)

# Check for duplicate rows
duplicate_rows = spotify_data.duplicated().sum()
print("\nNumber of duplicate rows:", duplicate_rows)

# drop rows with missing values
spotify_data_cleaned = spotify_data.dropna()

# Remove duplicate rows
spotify_data_cleaned = spotify_data_cleaned.drop_duplicates()

# Verify the cleaned data
missing_values_cleaned = spotify_data_cleaned.isnull().sum()
duplicate_rows_cleaned = spotify_data_cleaned.duplicated().sum()

print("\nMissing values after cleaning:\n", missing_values_cleaned)
print("\nNumber of duplicate rows after cleaning:", duplicate_rows_cleaned)

# Display first few rows of cleaned data
spotify_data_cleaned.head()


```

```python
#Q2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.


import pandas as pd
import matplotlib.pyplot as plt

print("Missing values per column:\n", missing_values)
print("\nNumber of duplicate rows:", duplicate_rows)
print("\nMissing values after cleaning:\n", missing_values_cleaned)
print("\nNumber of duplicate rows after cleaning:", duplicate_rows_cleaned)

# Plot a histogram of the Popularity column
plt.figure(figsize=(10, 6))
plt.hist(spotify_data_cleaned['Popularity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Popularity among Tracks')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.

import pandas as pd
import matplotlib.pyplot as plt


print("Missing values per column:\n", missing_values)
print("\nNumber of duplicate rows:", duplicate_rows)
print("\nMissing values after cleaning:\n", missing_values_cleaned)
print("\nNumber of duplicate rows after cleaning:", duplicate_rows_cleaned)

# Plot a scatter plot to explore the relationship between popularity and duration
plt.figure(figsize=(10, 6))
plt.scatter(spotify_data_cleaned['Duration (ms)'], spotify_data_cleaned['Popularity'], alpha=0.5, color='blue')
plt.title('Relationship between Track Duration and Popularity')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()


```

```python
#Q4.Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist
#    using a countplot.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Missing values per column:\n", missing_values)
print("\nNumber of duplicate rows:", duplicate_rows)
print("\nMissing values after cleaning:\n", missing_values_cleaned)
print("\nNumber of duplicate rows after cleaning:", duplicate_rows_cleaned)

# Count the number of tracks for each artist
artist_track_counts = spotify_data_cleaned['Artist'].value_counts()

# Display the artist with the highest number of tracks
top_artist = artist_track_counts.idxmax()
top_artist_count = artist_track_counts.max()
print(f"\nArtist with the highest number of tracks: {top_artist} ({top_artist_count} tracks)")

# Plot a countplot of the number of tracks for each artist
plt.figure(figsize=(35, 50))
sns.countplot(y='Artist', data=spotify_data_cleaned, order=spotify_data_cleaned['Artist'].value_counts().index, palette='viridis')
plt.title('Number of Tracks by Artist')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

```python

#Q5.What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.



# Find the top 5 least popular tracks
least_popular_tracks = spotify_data.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]

print(least_popular_tracks)
```

```python
#Q6.Among the top 5 most popular artists, which artist has the highest popularity on average?
#  Calculate and display the average popularity for each artist.


artist_avg_popularity = spotify_data.groupby('Artist')['Popularity'].mean().reset_index()

# 5 most popular artists based on average popularity
top_5_artists = artist_avg_popularity.nlargest(5, 'Popularity')

# 5 artists and their average popularity
print(top_5_artists)
```

```python
#Q7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.



top_5_artists = artist_avg_popularity.nlargest(5, 'Popularity')['Artist']

# most popular track for each of the top 5 artists
most_popular_tracks = spotify_data.loc[spotify_data.groupby('Artist')['Popularity'].idxmax()]

# only the top 5 artists
most_popular_tracks_top_5 = most_popular_tracks[most_popular_tracks['Artist'].isin(top_5_artists)]

# Select relevant columns
most_popular_tracks_top_5 = most_popular_tracks_top_5[['Artist', 'Track Name', 'Popularity']]

print(most_popular_tracks_top_5)
```

```python

#Q8.Visualize relationships between multiple numerical variables simultaneously using a pair plot.


import seaborn as sns
import matplotlib.pyplot as plt

# numerical columns for the pair plot
numerical_columns = ['Popularity', 'Duration (ms)']

# Create a pair plot
sns.pairplot(spotify_data[numerical_columns])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python

#Q9.Does the duration of tracks vary significantly across different artists?
# Explore this visually using a box plot or violin plot.



# For visualization purposes, we'll limit the number of artists to the top 10 most frequent artists
top_artists = spotify_data['Artist'].value_counts().nlargest(10).index
top_artists_data = spotify_data[spotify_data['Artist'].isin(top_artists)]

# Create a box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Artist', y='Duration (ms)', data=top_artists_data)
plt.title('Box Plot of Track Duration by Artist')
plt.xticks(rotation=90)
plt.show()

# Create a violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='Artist', y='Duration (ms)', data=top_artists_data)
plt.title('Violin Plot of Track Duration by Artist')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q10.How does the distribution of track popularity vary for different artists?
#  Visualize this using a swarm plot or a violin plot


# Create a swarm plot
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Artist', y='Popularity', data=top_artists_data)
plt.title('Swarm Plot of Track Popularity by Artist')
plt.xticks(rotation=90)
plt.show()

# Create a violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='Artist', y='Popularity', data=top_artists_data)
plt.title('Violin Plot of Track Popularity by Artist')
plt.xticks(rotation=90)
plt.show()
```
